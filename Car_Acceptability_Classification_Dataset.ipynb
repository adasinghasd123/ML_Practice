```python
X=df.iloc[:,0:6]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=8)
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder,OrdinalEncoder,LabelEncoder,MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import set_config
set_config(display='diagram')
from sklearn.model_selection import cross_val_score
from sklearn.impute import SimpleImputer
```

```python
X.info()
```

```python
preprocessor=ColumnTransformer([
    ('ordinal_Encoding1',OrdinalEncoder(categories=[['vhigh','high','med','low']]),[0]),
    ('ordinal_Encoding2',OrdinalEncoder(categories=[['vhigh','high','med','low']]),[1]),
    ('ordinal_Encoding3',OrdinalEncoder(categories=[['small','med','big']]),[4]),
    ('ordinal_Encoding4',OrdinalEncoder(categories=[['low','med','high']]),[5]),
    ('OHE',OneHotEncoder(sparse=False,handle_unknown='ignore'),[2,3])
])
```

```python
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
```

```python
X_train_fit=pipeline.fit_transform(X_train)
X_test_fit=pipeline.transform(X_test)
```

```python
X_train_fit[0]
```

```python
X_fit=pipeline.fit_transform(X)
```

```python
lr = LogisticRegression()
lr.fit(X_train_fit,y_train)
#print(np.mean(cross_val_score(lr,X,y,scoring='accuracy',cv=10)))



knn=KNeighborsClassifier()
knn.fit(X_train_fit,y_train)
#print(np.mean(cross_val_score(knn,X,y,scoring='accuracy',cv=10)))

dt=DecisionTreeClassifier()
dt.fit(X_train_fit,y_train)
#print(np.mean(cross_val_score(dt,X,y,scoring='accuracy',cv=10)))

rf=RandomForestClassifier()
rf.fit(X_train_fit,y_train)
#print(np.mean(cross_val_score(rf,X,y,scoring='accuracy',cv=10)))
```

```python
y_pred1 = lr.predict(X_test_fit)

y_pred3 = knn.predict(X_test_fit)
y_pred4 = dt.predict(X_test_fit)
y_pred5 = rf.predict(X_test_fit)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score1=accuracy_score(y_test,y_pred1)

score3=accuracy_score(y_test,y_pred3)
score4=accuracy_score(y_test,y_pred4)
score5=accuracy_score(y_test,y_pred5)
```

```python
print(score1*100,score3*100,score4*100,score5*100)
```

```python
print(np.mean(cross_val_score(lr,X_fit,y,scoring='accuracy',cv=10)))
print(np.mean(cross_val_score(knn,X_fit,y,scoring='accuracy',cv=10)))
print(np.mean(cross_val_score(dt,X_fit,y,scoring='accuracy',cv=10)))
print(np.mean(cross_val_score(rf,X_fit,y,scoring='accuracy',cv=10)))
```

**Since the accuracy of Decision Tree is maximum so now we are going to create the whole pipeline using DT**

```python
trf1=ColumnTransformer([
    ('ordinal_Encoding1',OrdinalEncoder(categories=[['vhigh','high','med','low']]),[0]),
    ('ordinal_Encoding2',OrdinalEncoder(categories=[['vhigh','high','med','low']]),[1]),
    ('ordinal_Encoding3',OrdinalEncoder(categories=[['small','med','big']]),[4]),
    ('ordinal_Encoding4',OrdinalEncoder(categories=[['low','med','high']]),[5]),
    ('OHE',OneHotEncoder(sparse=False,handle_unknown='ignore'),[2,3])
])
```

```python
trf2=DecisionTreeClassifier(criterion='entropy',
 max_depth=None,min_samples_leaf=2,min_samples_split= 2)
```

```python
pipe=Pipeline([
    ('trf1',trf1),
    ('trf2',trf2)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(np.mean(cross_val_score(pipe,X,y,scoring='accuracy',cv=10)))
```

```python
param_dist = {
    'trf2__criterion': ['gini', 'entropy'],
    'trf2__n_estimators': np.arange(100, 1000, 100),
    'trf2__max_depth': [None, 5, 10, 15],
    'trf2__min_samples_split': np.arange(2, 10),
    'trf2__min_samples_leaf': np.arange(1, 5),
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")
cv = GridSearchCV(pipe, param_grid=param_grid, cv=5)
cv.fit(X, y)
```
